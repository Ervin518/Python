```python
# The pyplot function makes some change to a figure: e.g., creates a figure,
# creates a plotting area in a figure, plots some lines in a plotting area,
# decorates the plot with labels, etc.
import matplotlib.pyplot as plt # pyplot package can be referred to as plt
import numpy as np  # the fundamental package for scientific computing with Python
```

# Linar plots



```python
x = [1, 2, 3, 4]  # list containing the points on the x-axis
y = np.array([10, 20, 30, 40]) # array containing the points on the y-axis
plt.plot(x, y) # The plot() function is used to draw points (markers) in a diagram.
plt.show()
```

```python
x = [1,2,3,4]
y = [10, 20, 30, 40]

plt.plot(x, y, color = 'red')
plt.title('This is 2*x for x = [1, 2, 3, 4]')
plt.xlabel('label of X')
plt.ylabel('label of Y')
plt.show()
```

**You can choose any of these markers:**
[link text](https://www.w3schools.com/python/matplotlib_markers.asp)

```python
x = [1,2,3,4]
y = [10, 20, 30, 40]   # the first point coordinate is (x1,y1) = (1,10)

plt.plot(x, y, marker = 'o')
plt.show()
```

```python
x = [1,2,3,4]
y = [10, 20, 30, 40]

plt.plot(x, y, 'r*')  #  'r*' ---> r = 'red' point color, '*' point style
plt.show()
```

```python
x = [1,2,3,4]
y = [10, 20, 30, 40]

plt.plot(x, y, 'g<')
plt.annotate('point', xy=(1.1, 10), xytext=(3, 10),
             arrowprops=dict(facecolor='black', shrink=0.01),
             )
plt.show()
```

## Several plots

```python
x1 = [1,2,3,4]
y1 = [10, 20, 30, 40]

x2 = [1,2,3,4]
y2 = [15, 25, 35, 45]

plt.plot(x1, y1, color = 'red', marker = '<')
plt.plot(x2, y2, color = 'green', marker = 's')
plt.show()
```

```python
import numpy as np

x = np.arange(0, 10, 0.5)  # min = 0, max = 10, steps = 0.5
# red dashes, blue squares and green triangles
plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^')
plt.show()
```

```python
x = np.linspace(1.0, 4.0, 90, endpoint=True) # x =<1;4> 90 points
y = x**2 - 5*x + 6
plt.plot(x, y)
plt.plot(x,0*x, color='pink')
# # 0*x - wektor złożony z zer o liczbie elementów równej zmiennej x
plt.plot([2,3], [0,0], 'o', color='green') # 2 points with coordinates (2,0),(3;0)
plt.grid(True)  # mesh
plt.show()
```

# Histograms


```python
# random numbers: 10000, “standard normal” distribution (sigma=1,  mean: 0)
x = np.random.randn(10000)
print(np.random.randn(10000))
```

```python

plt.hist(x, 100, facecolor ='blue') # 100 intervals
plt.xlabel('Intervals')
plt.ylabel('Likelihood')
plt.grid(True)
plt.show()
```

```python
miu=100
sigma = 20
x = miu + sigma*np.random.randn(10000)
plt.hist(x, 50, color = 'orange')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=20$')
plt.axis([40, 160, 0, 800])
plt.grid(True)
plt.show()
```

# Pie

```python
labels1 = ['Tulips', 'Roses', 'Lilies', 'Daffodils']
sizes = [25, 20, 40, 15]   # parts in percentages
explode1 = (0, 0., 0, 0)  # ejection of a specific part
fig1, ax1 = plt.subplots() # subplots - multiple charts in a one window
ax1.pie(sizes, explode=explode1, labels=labels1, autopct='%1.1f%%', shadow=True, startangle=45) # explode wysunięcie wykresu,
# startangle obrót wykresu
ax1.axis('equal')  # osie kwadratowe
plt.show()
```

## Plotting with categorical variables

```python
names = ['Dogs', 'Cats', 'Rats']
values = [5, 10, 15,]
plt.figure(figsize=(15, 5))  # width = 15, height = 5

plt.subplot(131)  # (131) 1 row, 3 columns, draw in 1 box
plt.bar(names, values)
plt.title('Dogs')
plt.subplot(132) # draw in 2 box
plt.scatter(names, values)
plt.subplot(133) # draw in 3 box
plt.plot(names, values)
plt.suptitle('Dogs and Cats and Rats')
plt.show()
```

Draw a scatter plot


*   color of markers: red
*   x,y range: from position (0,0) to position (5,25), 10 points
*   add labels to the x- and y-axis ('X', 'Y')
*   insert the coordinate (as text (0,0), (5,25)) in plot for the first and last point


```python
import numpy as np

x = np.arange(0, 5, 0.5)
y = np.arange(0, 25, 2.5)

plt.plot(x, y, 'r*', color = 'red', marker = 's')
plt.xlabel('X')
plt.ylabel('Y')
plt.text(x[0] + 0.2, y[0], '(0, 0)')
plt.text(x[-1] - 0.8, y[-1] - 1.5, '(5, 25)')
plt.show()
```
